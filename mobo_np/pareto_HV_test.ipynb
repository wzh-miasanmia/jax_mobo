```python
import numpy as np
from pareto_cal import is_non_dominated_np
from HV_cal import Hypervolume, plot_pareto_hv
```

check and return current pareto front

```python
Y = np.array([[1, 6], [2, 2], [1, 1], [2, 5], [3, 4], [5, 2], [4, 3]])
mask_no_dedup = is_non_dominated_np(Y)
# determine the number of points in Y that are Pareto optimal
num_new_pareto = mask_no_dedup[-Y.shape[-2] :].sum()
# non-dominated points
Y_nd = Y[mask_no_dedup]
print(num_new_pareto, Y_nd)
```

calculate Hypervolume of the pareto front points

```python
ref_point = np.array([0,0])
HV = Hypervolume(ref_point)
HV_value = HV.compute(Y_nd)
print(HV_value)
```

visualization

```python
plot_pareto_hv(Y_nd, ref_point)
```
